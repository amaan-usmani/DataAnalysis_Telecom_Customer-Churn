```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


telecom_df = pd.read_csv("C:\\Users\\amaan\\Downloads\\customer_churn project Database.csv")
telecom_df.drop(["StreamingMovies","PaperlessBilling"], axis = 1, inplace = True)
telecom_df
```

```python
# Variable assigning to Churned data

churn_df = telecom_df[telecom_df["Churn"]=="Yes"]
churn_df
```

```python
# tenure wise customer churning 

tenure_cust_churn = churn_df.groupby(['tenure'])['Churn'].value_counts().reset_index(name="churn_customer")
print(tenure_cust_churn.sum())
tenure_cust_churn 
```

```python
# Total no.of Customers Churning with Phone Service

churn_phn_yes = churn_df[churn_df["PhoneService"]=="Yes"]
print(f"No.of Customers churning having Phone Service are {churn_phn_yes['PhoneService'].value_counts()[0]}")   

# Total no. of Customers Churning without Phone Service

churn_phn_no = churn_df[churn_df["PhoneService"]=="No"]
print(f"No.of Customers churning not having Phone Service are {churn_phn_no['PhoneService'].value_counts()[0]}")                     
```

```python
# No. of Customer Churining with Internet service as DSL

churn_inter_DSL = churn_df[churn_df["InternetService"]=="DSL"]
print(f"No. of Customer Churning having Internet Service as DSL are {churn_inter_DSL['InternetService'].value_counts()[0]}")

# No. of Customer Churining with Internet service as Fiber optic

churn_inter_DSL = churn_df[churn_df["InternetService"]=="Fiber optic"]
print(f"No. of Customer Churning having Internet Service as Fiber optic are {churn_inter_DSL['InternetService'].value_counts()[0]}")
```

```python
# Cobining data of Phone Service and Internet Service

# Customers churning Phone service and Internet service as DSL

phn_service_DSL = churn_df[(churn_df["PhoneService"]=="Yes")&(churn_df["InternetService"]=="DSL")]
print(f"No.of Customers churning having Phone Service and Internet Service as DSL are {phn_service_DSL['Churn'].value_counts()}")

# Customers churning phone service and Internet Service as Fiber optic

phn_service_FO = churn_df[(churn_df["PhoneService"]=="Yes")&(churn_df["InternetService"]=="Fiber optic")]
print(f"No.of Customers churning having Phone Service and Internet Service as Fiber optic are {phn_service_DSL['Churn'].value_counts()}")

#customers churning without Phone service but with Internet service as DSL

NO_phn_service_DSL = churn_df[(churn_df["PhoneService"]=="No")&(churn_df["InternetService"]=="DSL")]
print(f"No.of Customers churning not having Phone Service but having Internet Service as DSL are {phn_service_DSL['Churn'].value_counts()}")
```

```python
fig, axes = plt.subplots(4, 4, figsize=(20, 16), constrained_layout=True)

# Iterate over predictors and plot each one
for i, predictor in enumerate(telecom_df.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges', 'tenure'])):
    row = i // 4
    col = i % 4
    ax = axes[row, col]
    sns.countplot(data=telecom_df, x=predictor, hue='Churn', ax=ax, palette=['skyblue', 'salmon'])
    ax.set_title(predictor)

plt.tight_layout()
plt.show()
```
